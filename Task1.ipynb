<h1>18.BMI category</h1>

```python
def bmi_category(bmi):
    if bmi<=18.5:
        return 'Underweight'
    elif bmi>18.5 and bmi<=25:
        return 'Normal'
    elif bmi>25 and bmi<=35:
        return 'Overweight'
    elif bmi>35:
        return 'obesity'

df['bmi_category'] = df['bmi'].apply(bmi_category)
df['bmi_category']
```

<h1> Implement one-hot encoding using pd.get_dummies()</h1>

```python
Categorical_feature = "bmi_category"
df_encoded = pd.get_dummies(df, columns = [Categorical_feature], drop_first = True)
```

<h1>Implementing Z_score scaling for numerical features</h1>

```python
Numerical_feature = ['height', 'weight', 'ap_hi', 'ap_lo', 'age_years', 'bmi']
def z_score_scale(series):
    mean = series.mean()
    std = series.std()
    return (series - mean)/std
```

<h1>Apply scaling function to numerical columns</h1>

```python
for col in Numerical_feature:
    df_encoded[col + '_scaled'] = z_score_scale(df_encoded[col])
```

<h1>Drop the original unscaled numerical columns</h1>

```python
df_processed = df_encoded.drop(columns=Numerical_feature)
```

<h1>Saving final processed data</h1>

```python
df_processed.to_csv('cardio_processed.csv', index = False)
```
